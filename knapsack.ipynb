```python
# 0/1 knapsack using dynamic programing

import numpy as np
import time

def knapsack(capacity, weights, values, no_of_items):
    n = no_of_items
    # initializing the dp[][] array with 0s
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    if(weights==None or values==None or len(weights)!=len(values) or capacity < 0):
        raise ValueError("Invalid argument, check for input items !!!")
    

    # Fill the dp table
    start_time = time.time()
    for i in range(1, n + 1):
        for w in range(1,capacity + 1):
            if weights[i - 1] > w:
                dp[i][w] = dp[i - 1][w]
            else:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    
    # Now, dp[n][capacity] contains the optimal value
    optimal_value = dp[n][capacity]
    

    # Trace back to find the included items
    included = []      # Selected items intex list
    tflist = []        # Item list of True/False
    w = capacity
    total_weight = 0
    count=0
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i - 1][w]:   #1D lists so, i-1 is needed
            included.append(i - 1)    #(i-1)because our lists index starts from 0
            count = count + 1   # so, n=5 means weights[0 to 4]
            tflist.append('TRUE')
        else:
            tflist.append('FALSE')
        
    used_capacity = total_weight
    available_capacity = capacity - used_capacity
    total_weight += weights[i-1]
    w -= weights[i - 1]    
    return optimal_value, tflist, included, elapsed_time, used_capacity,  available_capacity, count

# Example usage:
no_of_items = 140
w = np.random.randint(1,21,no_of_items)
v = w + 5
capacity = int(0.5 * np.sum(w))
weights = w.tolist()
values = v.tolist()

if __name__ == "__main__":
   print("WEIGHTS = ",weights,"\n")
   print("VALUES = ",values,"\n")
   print("CAPASITY = ",capacity)
   optimal_value, tflist, included, elapsed_time, used_capacity,  available_capacity, count = knapsack(capacity, weights, values, no_of_items)
   print('='*70)
   print(f"Used knapsack capacity: {used_capacity}\n")
   print(f"Avalable knapsack capacity: {available_capacity}\n")
   print(f"Included item no. count: {count}\n")
   print(f"Optimal knapsack value: {optimal_value}\n")
   print(f"Time = {elapsed_time}\n")
   print(f"Items index for included items: {list(reversed(included))}\n")
   print(f"Selected items: {list(reversed(tflist))}\n")
```
